```python
from selenium import webdriver
import re
driver = webdriver.Chrome('../studyFile/chromedriver.exe')
```

```python
str_list = ["정치","경제","사회","생활/문화","세계","IT/과학"]
url = "https://news.naver.com/main/main.nhn?mode=LSD&mid=shm&sid1=10"
res_list = []
for i in range(len(str_list)):
    tmp_list=[]
    driver.get(url+str(i))
    
    links = driver.find_elements_by_css_selector("div[class=cluster_text] > a")
    #print(links)
    #tmp_list.extend([link.text for link in links])
    for link in links:
        tmp_list.append(re.sub("[^ㄱ-ㅎㅏ-ㅣ가-힣]","",link.text))
    res_list.append(tmp_list)
    
```

```python
from matplotlib import font_manager,rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font',family=font_name)
```

```python
import nltk
from konlpy.tag import Kkma
k = Kkma()
```

```python
for i,res in enumerate(res_list):
    nouns = [k.morphs(i) for i in res]
    tmp=[]
    for noun in nouns:
        tmp+=noun
    koNouns = nltk.Text(tmp)
    print(str_list[i]+'면 주요 10개 키워드')
    koNouns.plot(10)
    
```
